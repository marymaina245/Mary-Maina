**Seasonal Decompostion**
Seasonal decomposition is a statistical method used to break down a time series into its core components: trend, seasonality, and residual (or noise). The trend captures the long-term direction of the data, such as a consistent increase or decrease over time. Seasonality identifies repeating patterns or cycles within a fixed period, like yearly fluctuations in stock prices due to annual market trends. The residual represents the remaining irregular or random variations after removing the trend and seasonal effects.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Decompose the time series for MasterCard
decomposition_M = seasonal_decompose(df['Close_M'], model='multiplicative', period=365)
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 12))

ax1.plot(decomposition_M.observed)
ax1.set_title('Observed - MasterCard')
ax2.plot(decomposition_M.trend)
ax2.set_title('Trend - MasterCard')
ax3.plot(decomposition_M.seasonal)
ax3.set_title('Seasonal - MasterCard')
ax4.plot(decomposition_M.resid)
ax4.set_title('Residual - MasterCard')

plt.tight_layout()
plt.show()
```

The above charts show trend,seaonality and residual components of Mastercard.The residuals indicate noise which need preprocessing to improve forecasting accuracy.The trend help in identifying key patterns

\ **Check for stationarity using the Augmented Dickey-Fuller test**

The Augmented Dickey-Fuller (ADF) test is a statistical test used to determine whether a time series is stationary, meaning its statistical properties like mean and variance are constant over time. It tests the null hypothesis that the time series has a unit root (i.e., it is non-stationary) against the alternative that it is stationary. The test produces an ADF statistic, which is compared to critical values at different significance levels (e.g., 1%, 5%, 10%), and a p-value, where a p-value below 0.05 typically rejects the null hypothesis, indicating stationarity

```python
from statsmodels.tsa.stattools import adfuller

# Check for stationarity using the Augmented Dickey-Fuller test
def adf_test(series):
    result = adfuller(series.dropna())
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    for key, value in result[4].items():
        print('Critial Values:')
        print(f'   {key}, {value}')

print("ADF Test for MasterCard Close Price:")
adf_test(df['Close_M'])

print("\nADF Test for Visa Close Price:")
adf_test(df['Close_V'])
```

The Augmented Dickey-Fuller (ADF) test results for Mastercard and Visa closing prices indicate that both time series are non-stationary, as their ADF statistics (0.724 for Mastercard and 0.598 for Visa) are greater than the critical values at 1%, 5%, and 10% significance levels, and their p-values (0.990 and 0.988, respectively) are far above 0.05, failing to reject the null hypothesis of a unit root. This non-stationarity, observed over the 20082024 period, suggests that the stock prices have trends and are not mean-reverting   From the results models like ARIMA will require differencing or detrending to achieve seasonality.LSTM models can handle non-stationary data

Differencing the data and checking the stationarity of the differenced series again

```python

# Select Close_M and Close_V for time series analysis
ts_m = df['Close_M']
ts_v = df['Close_V']

# Function to perform ADF test
def adf_test(series, title='df'):
    result = adfuller(series.dropna())
    print(f'ADF Test for {title}:')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'   {key}: {value}')
    if result[1] < 0.05:
        print('Result: Stationary (reject null hypothesis)\n')
    else:
        print('Result: Non-stationary (fail to reject null hypothesis)\n')

# Perform ADF test for both stocks
adf_test(ts_m, 'Mastercard Close_M')
adf_test(ts_v, 'Visa Close_V')

# Plot original series for both stocks
plt.figure(figsize=(12, 6))
plt.plot(ts_m, label='Mastercard Close_M', color='blue')
plt.plot(ts_v, label='Visa Close_V', color='orange')
plt.title('Mastercard and Visa Closing Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Difference the series for both stocks
ts_m_diff = ts_m.diff().dropna()
ts_v_diff = ts_v.diff().dropna()

# Perform ADF test on differenced series
adf_test(ts_m_diff, 'Differenced Mastercard Close_M')
adf_test(ts_v_diff, 'Differenced Visa Close_V')

# Plot differenced series
plt.figure(figsize=(12, 6))
plt.plot(ts_m_diff, label='Differenced Mastercard Close_M', color='blue')
plt.plot(ts_v_diff, label='Differenced Visa Close_V', color='orange')
plt.title('Differenced Closing Prices for Mastercard and Visa')
plt.xlabel('Date')
plt.ylabel('Price Difference')
plt.legend()
plt.show()
```

The series is now stationary and arima model can now be fitted

**Fitting the ARIMA model** with order(1,1,1)

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf

# Define function to fit ARIMA, predict, evaluate, and analyze residuals
def fit_arima(series, order, title='df'):
    # Split into train and test sets
    train_size = int(len(series) * 0.8)  # 80% training, 20% testing
    train, test = series[:train_size], series[train_size:]

    # Fit ARIMA model
    model = ARIMA(train, order=order)
    model_fit = model.fit()  # Use default 'statespace' method for optimization
    print(f"{title} ARIMA{order} Summary:")
    print(model_fit.summary())

    # Predict on test set
    predictions = model_fit.forecast(steps=len(test))

    # Plot training, actual, and predicted values
    plt.figure(figsize=(12, 6))
    plt.plot(train.index, train, label='Training Data', color='blue')
    plt.plot(test.index, test, label='Actual Data', color='green')
    plt.plot(test.index, predictions, label='Predicted Data', color='red', linestyle='--')
    plt.title(f'{title} - ARIMA{order} Predictions')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Evaluate
    # Convert to NumPy arrays to avoid index alignment issues
    test_values = test.values
    predictions_values = predictions.values

    mse = mean_squared_error(test_values, predictions_values)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(test_values, predictions_values)
    # Calculate MAPE, handling zero values by masking
    mask = test_values != 0
    mape = np.mean(np.abs((test_values[mask] - predictions_values[mask]) / test_values[mask])) * 100 if mask.any() else np.nan

    print(f"Performance for {title}:")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.4f}%")
    print()

    # Residual analysis
    residuals = model_fit.resid
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(residuals.index, residuals, label='Residuals', color='purple')
    plt.title(f'{title} - ARIMA{order} Residuals')
    plt.xlabel('Date')
    plt.ylabel('Residual')
    plt.legend()
    plt.grid(True)

    plt.subplot(2, 1, 2)
    plot_acf(residuals, lags=40, ax=plt.gca())
    plt.title(f'ACF of {title} Residuals')

    plt.tight_layout()
    plt.show()

    return model_fit

# Fit ARIMA(1,1,1) models for both stocks
order = (1, 1, 1)  # Use ARIMA(1,1,1) based on ACF/PACF
model_m = fit_arima(ts_m, order, 'Mastercard Close_M')
model_v = fit_arima(ts_v, order, 'Visa Close_V')
```

Forecasting future stock prices

```python
from statsmodels.tsa.arima.model import ARIMA
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Retrain ARIMA(1,1,1) models on the full dataset
order = (1, 1, 1)

# Fit ARIMA model for Mastercard on the full dataset
model_m = ARIMA(ts_m, order=order)
model_fit_m = model_m.fit()
print("Mastercard ARIMA(1,1,1) Summary (Full Data):")
print(model_fit_m.summary())

# Fit ARIMA model for Visa on the full dataset
model_v = ARIMA(ts_v, order=order)
model_fit_v = model_v.fit()
print("\nVisa ARIMA(1,1,1) Summary (Full Data):")
print(model_fit_v.summary())

# Step 2: Generate future dates for forecasting (July 2024 to December 2027)
last_date = ts_m.index[-1]  # Last date in historical data (June 2024)
forecast_steps = 882  # Approx. trading days from July 2024 to Dec 2027
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=forecast_steps, freq='B')

# Step 3: Forecast future values
forecast_m = model_fit_m.forecast(steps=forecast_steps)
forecast_v = model_fit_v.forecast(steps=forecast_steps)

# Convert forecasts to pandas Series with future dates as index
forecast_m_series = pd.Series(forecast_m.values, index=future_dates)
forecast_v_series = pd.Series(forecast_v.values, index=future_dates)

# Step 4: Plot historical and forecasted data together
plt.figure(figsize=(12, 6))
# Mastercard: Historical and Forecast
plt.plot(ts_m.index, ts_m, label='Mastercard Historical', color='blue')
plt.plot(forecast_m_series.index, forecast_m_series, label='Mastercard Forecasted', color='darkblue', linestyle='--')
# Visa: Historical and Forecast
plt.plot(ts_v.index, ts_v, label='Visa Historical', color='orange')
plt.plot(forecast_v_series.index, forecast_v_series, label='Visa Forecasted', color='darkorange', linestyle='--')
plt.title('Mastercard and Visa Stock Prices: Historical and Forecasted (to 2027)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Calculate percentage returns for both stocks
returns_m = ts_m.pct_change().dropna() * 100
returns_v = ts_v.pct_change().dropna() * 100

# Fit GARCH(1,1) model for Mastercard
garch_model_m = arch_model(returns_m, vol='Garch', p=1, q=1, dist='Normal')
garch_fit_m = garch_model_m.fit(disp='off')
print("Mastercard GARCH(1,1) Summary:")
print(garch_fit_m.summary())

# Fit GARCH(1,1) model for Visa
garch_model_v = arch_model(returns_v, vol='Garch', p=1, q=1, dist='Normal')
garch_fit_v = garch_model_v.fit(disp='off')
print("\nVisa GARCH(1,1) Summary:")
print(garch_fit_v.summary())

# Extract conditional volatility
cond_vol_m = garch_fit_m.conditional_volatility
cond_vol_v = garch_fit_v.conditional_volatility

# Plot conditional volatility
plt.figure(figsize=(12, 6))
plt.plot(returns_m.index, cond_vol_m, label='Mastercard Volatility', color='blue')
plt.plot(returns_v.index, cond_vol_v, label='Visa Volatility', color='orange')
plt.title('GARCH(1,1) Conditional Volatility for Mastercard and Visa')
plt.xlabel('Date')
plt.ylabel('Volatility (%)')
plt.legend()
plt.savefig('garch_volatility_both.png')
plt.show()
```

**Correlation heatmap between the variables**
A correlation matrix is a table that shows the correlation coefficients between multiple variables in a dataset. Each cell in the matrix represents the strength and direction of the relationship between two variables, typically ranging from -1 to 1. A value of 1 indicates a perfect positive correlation (variables move together), -1 indicates a perfect negative correlation (variables move in opposite directions), and 0 indicates no correlation.


```python
# Select only numeric columns for correlation
numeric_cols = ['Open_M', 'High_M', 'Low_M', 'Close_M', 'Adj Close_M', 'Volume_M',
                'Open_V', 'High_V', 'Low_V', 'Close_V', 'Adj Close_V', 'Volume_V']
 # Create correlation matrix
correlation_matrix = df[numeric_cols].corr()
 # Visualize the correlation matrix using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

The correlation matrix reveals strong relationships between stock price metrics for two datasets, labeled "M" and "V." Open, High, Low, Close, and Adjusted Close prices within each dataset (e.g., Open_M, High_M, etc.) are nearly perfectly correlated (1.00), indicating that these price metrics move in lockstep for both "M" and "V." However, the price metrics between "M" and "V" (e.g., Open_M vs. Open_V) also show a very high positive correlation (1.00), suggesting that the two datasets likely track similar or highly related assets. Volume, on the other hand, shows a moderate negative correlation with price metrics within each dataset (around -0.41 to -0.44), implying that higher trading volumes might be associated with price declines. Interestingly, Volume_M and Volume_V have a strong positive correlation (0.78), indicating that trading volumes for both datasets tend to rise and fall together. This suggests that while the price movements of "M" and "V" are tightly linked, volume dynamics provide a nuanced contrast, potentially reflecting market behavior or external factors affecting both datasets similarly.
